```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('Waiting_times.xlsx')
df.head()   
```

```python
import matplotlib.pyplot as plt
plt.bar(df['Avg waiting time'], df['Year'])
plt.bar(df['Avg waiting time'], df['Year'],bottom=df['Year'])
plt.xlabel('Year')
plt.ylabel('Waiting time')
plt.title('Waiting time in the UK')


plt.show()
```

```python
df.plot(x='Month', y=['Min', 'Avg waiting time', 'Max'], kind='line', marker='o')
plt.xlabel('Month')
plt.ylabel('Waiting time')
plt.title('Waiting time in the UK')
plt.legend(['Min', 'Avg waiting time', 'Max'])
plt.show()
```

```python
df = pd.read_excel('Week 4 Exercise 2 - RIVM data.xlsx')
df.head()
```

```python
Weeks = df[['wk 33', 'wk 34', 'wk 35', 'wk 36', 'wk 37', 'wk 38']]
Age_groups = df['Age group']
plt.plot(Age_groups,Weeks)
plt.xlabel('Age group')
plt.ylabel('Weeks')
plt.title('Weekly Data')
plt.legend(['wk 33', 'wk 34', 'wk 35', 'wk 36', 'wk 37', 'wk 38'])
plt.show()
```

```python
Weeks = df[['wk 33', 'wk 34', 'wk 35', 'wk 36', 'wk 37', 'wk 38']]
Age_groups = df['Age group']
Weeks.plot(kind='bar', stacked=True)
plt.xlabel('Age group')
plt.ylabel('Weeks')
plt.title('Weekly Data by Age Group')
plt.xticks(ticks=range(len(Age_groups)), labels=Age_groups, rotation=45)
plt.legend(title='Weeks')
plt.show()
```

```python
import seaborn as sns

# Set the size of the heatmap
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(df.set_index('Age group').T, annot=True, cmap='YlGnBu')

# Set the title and labels
plt.title('Heatmap of Weekly Data by Age Group')
plt.xlabel('Age group')
plt.ylabel('Weeks')

# Show the plot
plt.show()
```
